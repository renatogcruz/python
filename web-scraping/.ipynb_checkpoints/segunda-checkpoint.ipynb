```python
# Importando bibliotecas
from urllib.request import urlopen, urlretrieve, Request
from bs4 import BeautifulSoup
import pandas as pd

# Função para obter HTML
def obtendoHtml(link):
    response = urlopen(link)
    html = response.read().decode('utf-8')
    soup = BeautifulSoup(html, 'html.parser')

# Declarando variável cards
cards = []

# Obtendo o HTML
obtendoHtml('https://www.euamomomi.com.br/blusa-infantil-feminina')
#response = urlopen('https://www.euamomomi.com.br/blusa-infantil-feminina')
#html = response.read().decode('utf-8')
#soup = BeautifulSoup(html, 'html.parser')


url = 'https://www.euamomomi.com.br/blusa-infantil-feminina'
headers = {'User-Agent': 'Mozilla/5.0 (Windows NT 10.0; Win64; x64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/92.0.4515.107 Safari/537.36'}
req = Request(url, headers = headers)
response = urlopen(req)
html = response.read().decode('utf-8')
soup = BeautifulSoup(html, 'html.parser')


# Obtendo as TAGs de interesse
anuncios = soup.find('div', {"id": "lista-produtos-area"}).findAll('li')

# Coletando as informações dos cards
for anuncio in anuncios:
    card = {}
    
    # nome do produto
    if anuncio.find('h3', {'class': 'product-name'}) != None:
        card['nome do produto'] = anuncio.find('h3', {'class': 'product-name'}).get_text()
                
        
    # tamanhos (lista com tamanhos disponíveis)
    if anuncio.find('div', {'class': 'derivacao-lista'}) != None:
        tamanhos = anuncio.find('div', {'class': 'derivacao-lista'}).findAll('div', {'class': 'deriva tamanho'})
            
        x = []

        for tamanho in tamanhos:
            num = tamanho.get_text().replace(' ', '')
            x.append(num)
        card['tamanhos'] = x

    
    # Obtendo hiperlink
    #hiperlink = anuncio.find('a')
    if anuncio.find('a') != None:
        link = 'https:' + anuncio.find('a').get('href')
    
                
        # chamando a função novamente
        #obtendoHtml(link)
        resp = urlopen(link)
        html2 = resp.read().decode('utf-8')
        soup2 = BeautifulSoup(html2, 'html.parser')
        dados = soup2.find('div', {'class': 'lista-dados-produto'})
     
        
        # descrição do produto
        descricao = dados.find('div', {'class': 'content'})
        # TESTE
        card['Teste'] = descricao
        
        #if descricao != None:
        #    card['descrição do produto'] = descricao.get_text()
    
    
        #  composição
        #composicao = dados.find('div', {'class': 'composicao'}).find('div', {'class': 'content'})
        #if composicao != None:
        #    card['Composição'] = composicao.get_text()
    
            
        
    # fotos (lista com as urls das imagens)
    image = anuncio.find('img', {'class': 'img-principal'})
    if image != None:
        card['fotos'] = image.get('data-imagem-principal')       
        

    # Adicionando resultado à lista cards

    cards.append(card)


# Criando um DataFrame com os resultados
dataset = pd.DataFrame(cards)
#dataset.to_csv('./output/data/dataset.csv', sep=';', index = False, encoding = 'utf-8-sig')
dataset
```

# Criando uma rotina de scraping

```python
# Importando bibliotecas
from urllib.request import urlopen, urlretrieve, Request
from bs4 import BeautifulSoup
import pandas as pd

# Função para obter HTML

def obtendoHtml(link):
    response = urlopen(link)
    html = response.read().decode('utf-8')
    soup = BeautifulSoup(html, 'html.parser')

# Declarando variável cards
cards = []

# Obtendo o HTML

obtendoHtml('https://www.euamomomi.com.br/blusa-infantil-feminina')
#response = urlopen('https://www.euamomomi.com.br/blusa-infantil-feminina')
#html = response.read().decode('utf-8')
#soup = BeautifulSoup(html, 'html.parser')

# Obtendo as TAGs de interesse
anuncios = soup.find('div', {"id": "lista-produtos-area"}).findAll('li')

# Coletando as informações dos cards
for anuncio in anuncios:
    card = {}
    
    # Nome
    if anuncio.find('h3', {'class': 'product-name'}) != None:
        card['Produto'] = anuncio.find('h3', {'class': 'product-name'}).get_text()
                
        
    # Tamanhos
    if anuncio.find('div', {'class': 'derivacao-lista'}) != None:
        tamanhos = anuncio.find('div', {'class': 'derivacao-lista'}).findAll('div', {'class': 'deriva tamanho'})
            
        x = []

        for tamanho in tamanhos:
            num = tamanho.get_text().replace(' ', '')
            x.append(num)
        card['Tamanho'] = x
       
        
        
    # Imagem
    image = anuncio.find('img', {'class': 'img-principal'})
    if image != None:
        card['Imagens'] = image.get('data-imagem-principal')       
        

    # Adicionando resultado à lista cards

    cards.append(card)


# Criando um DataFrame com os resultados
dataset = pd.DataFrame(cards)
#dataset.to_csv('./output/data/dataset.csv', sep=';', index = False, encoding = 'utf-8-sig')
dataset
```
